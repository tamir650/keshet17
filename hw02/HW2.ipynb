```python
city_list = list()
forest_list = list()

for i in range(10):
    img = mpimg.imread('../data/forest' + str(i) + '.jpg')
    forest_list.append(img[:,:,0].sum() / img[:,:,0].size)

    img = mpimg.imread('../data/city' + str(i) + '.jpg')
    city_list.append(img[:,:,0].sum() / img[:,:,0].size)


```

```python
city_list    
```

```python
forest_list
```

```python
forest_array = np.array(forest_list)
plt.plot(forest_array, np.zeros(forest_array.shape), 'r+')

city_array = np.array(city_list)
plt.plot(city_array, np.zeros(city_array.shape), 'bo')

plt.show()
```
