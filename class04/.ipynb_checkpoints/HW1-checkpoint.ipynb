```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img = mpimg.imread('forest2.jpg')
plt.imshow(img)
```

```python
green_image = img.copy() # Make a copy
green_image[:,:,0] = 0
green_image[:,:,2] = 0 
plt.imshow(green_image)
```

```python
red_image = img.copy() # Make a copy
red_image[:,:,1] = 0
red_image[:,:,2] = 0
plt.imshow(red_image)
```

```python
blue_image = img.copy() # Make a copy
blue_image[:,:,0] = 0
blue_image[:,:,1] = 0
plt.imshow(blue_image)
```

```python
city_list = list()
forest_list = list()

for i in range(10):
    img = mpimg.imread('forest' + str(i) + '.jpg')
    forest_list.append(img[:,:,0].sum() / img[:,:,0].size)

    img = mpimg.imread('city' + str(i) + '.jpg')
    city_list.append(img[:,:,0].sum() / img[:,:,0].size)


```

```python
city_list    
```

```python
forest_list
```

```python
forest_array = np.array(forest_list)
plt.plot(forest_array, np.zeros(forest_array.shape), 'r+')

city_array = np.array(city_list)
plt.plot(city_array, np.zeros(city_array.shape), 'bo')

plt.show()
```

```python
city_list = list()
forest_list = list()

for i in range(10):
    img = mpimg.imread('forest' + str(i) + '.jpg')
    forest_list.append([img[:,:,0].sum() / img[:,:,0].size, img[:,:,2].sum() / img[:,:,2].size])

    img = mpimg.imread('city' + str(i) + '.jpg')
    city_list.append([img[:,:,0].sum() / img[:,:,0].size, img[:,:,2].sum() / img[:,:,2].size])



```

```python
city_list    
```

```python
forest_list
```

```python
forest_array = np.array(forest_list)
plt.plot(forest_array[:,0], forest_array[:,1], 'r+')

city_array = np.array(city_list)
plt.plot(city_array[:,0], city_array[:,1], 'bo')

plt.plot([40, 142], [155, 25], 'k-', lw=2)

plt.show()
```
